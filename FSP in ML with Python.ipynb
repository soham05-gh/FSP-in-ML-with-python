1. Write a Python program to create a DataFrame containing missing values. Perform the following operations: • Detect missing values  • Replace missing values with mean/median  • Drop rows containing missing values 

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({ 'Age': [25, 30, np.nan, 35, np.nan], 'Salary': [40000, np.nan, 45000, 50000, 55000] })
print(df)

print(df.isnull())

df_mean = df.fillna(df.mean(numeric_only=True))
print("\nFilled with Mean:")
print(df_mean)

df_median = df.fillna(df.median(numeric_only=True))
print("\nFilled with Median:")
print(df_median)

df_drop = df.dropna()
print("\nRows after dropping missing values:")
print(df_drop)
```

2. Write a Python program using Pandas to: • Create a dataset containing duplicate rows  • Identify duplicate entries  • Remove all duplicate rows from the dataset 

```python
import pandas as pd
df = pd.DataFrame({ 'Name': ['John', 'Bob', 'Charles', 'John', 'Bob'], 'Age': [35, 25, 30, 35, 25] })
print("Original Dataset:")
print(df)

print("\nDuplicate Rows:")
print(df.duplicated())

df_unique = df.drop_duplicates()
print("\nDataset after duplicate removal:")
print(df_unique)
```

3. Write a Python program to normalize the following list of values using Min-Max normalization: [12, 25, 40, 55, 70] • Use the formula: • 𝑥′ = 𝑥−𝑥𝑚𝑖𝑛/𝑥𝑚𝑎𝑥−𝑥𝑚𝑖𝑛

```python
data = [12, 25, 40, 55, 70]

xmin = min(data)
xmax = max(data)

normalized = [(x - xmin)/(xmax - xmin) for x in data]

print("Original Data:", data)
print("Normalized Data:", normalized)
```

4. Write a Python program to standardize a dataset using Z-score normalization. The program should: • Calculate mean and standard deviation  • Compute Z-score for each value  • Display transformed data 

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({ 'Values': [10, 25, 40, 50, 75] })

print("Original DataFrame:")
print(df)

mean = df['Values'].mean()
std = df['Values'].std()

print("\nMean =", mean)
print("Standard Deviation =", std)

df['Z_Score'] = (df['Values'] - mean)/std

print("\nStandardized Data:")
print(df)
```

5. Write a Python program to detect outliers in a dataset using the Z-score method. 

```python
import numpy as np
import pandas as pd
from scipy.stats import zscore

df = pd.DataFrame({ 'Marks': [10, 12, 14, 16, 18, 20, 75]})
print("Original DataFrame:")
print(df)

df['Z_Score'] = zscore(df['Marks'])
print("\nDataFrame with Z-Scores:")
print(df)

outliers = df[np.abs(df['Z_Score'])>2]
print("\nOutliers:")
print(outliers)
```

6. Write a Python program to: • Create a dataset containing categorical values such as ["Red", "Blue", "Green", "Blue"]  • Convert the categorical data into numerical labels using Label Encoding 

```python
from sklearn.preprocessing import LabelEncoder

colors = ["Red", "Blue", "Green", "Blue"]

encoder = LabelEncoder()
encoded = encoder.fit_transform(colors)

print("Original:", colors)
print("Encoded:", encoded)
```

7. Write a Python program to perform One-Hot Encoding on a categorical feature named City containing values: Kolkata, Delhi, Mumbai, Chennai 

```python
import pandas as pd
df = pd.DataFrame({ 'City': ['Kolkata', 'Delhi' , 'Mumbai', 'Chennai'] })

encoded_df = pd.get_dummies(df, columns=['City'])
print(encoded_df)
```

8. Write a Python program using StandardScaler from Scikit-learn to scale numerical features of a dataset. 

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

df = pd.DataFrame({ 'Marks': [40, 55, 60, 75, 90] })
print(df)

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

print("Scaled Data:")
print(scaled_data)
```

9. Write a Python program to compare: • Min-Max Scaling  • Standardization  on the same dataset and display the transformed outputs. 

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd

df = pd.DataFrame({ 'Values': [15, 25, 45, 65, 80] })

minmax = MinMaxScaler()
minmax_scaled = minmax.fit_transform(df)

standard = StandardScaler()
standard_scaled = standard.fit_transform(df)

print("Original Data:")
print(df)

print("Min-Max Scaled:")
print(minmax_scaled)

print("Standardized:")
print(standard_scaled)
```

10. Write a Python program to convert a column containing numeric strings into integer type. 

```python
import pandas as pd

df = pd.DataFrame({ 'Numbers': ['10', '20', '30', '45', '60'] })

print("Before Conversion:")
print(df.dtypes)

df['Numbers'] = df['Numbers'].astype(int)

print("\nAfter Conversion:")
print(df.dtypes)

print("\nDataFrame:")
print(df)
```
